## Código Base

````python
import pandas as pd
import numpy as np
from typing import Dict, List, Tuple, Optional
````

````python
def estadisticas_basicas(precios: pd.Series) -> Dict:
    """
    Calcula estadísticas descriptivas de los precios.
    """
    resultado = {
        "precio_actual": precios.iloc[-1],
        "precio_minimo": precios.min(),
        "precio_maximo": precios.max(),
        "precio_promedio": precios.mean(),
        "precio_mediana": precios.median(),
        "desviacion_std": precios.std(),
        "rango": precios.max() - precios.min(),
        "dias_analizados": len(precios)
    }
    
    return resultado
````

````python
def calcular_rendimientos(precios: pd.Series) -> pd.Series:
    """
    Calcula el rendimiento diario porcentual.
    """
    return precios.pct_change() * 100
````

````python
def analisis_rendimientos(rendimientos: pd.Series) -> Dict:
    """
    Analiza los rendimientos calculados.
    """
    resultado = {
        "rendimiento_total": None,
        "rendimiento_promedio": None,
        "mejor_dia": None,
        "peor_dia": None,
        "dias_positivos": None,
        "dias_negativos": None,
        "volatilidad": None
    }
    
    # Excluimos el primer día (NaN, no tiene día anterior para comparar)
    rend_validos = rendimientos.dropna()
    
    if rend_validos.empty:
        resultado.update({
            "rendimiento_total": 0.0,
            "rendimiento_promedio": 0.0,
            "dias_positivos": 0,
            "dias_negativos": 0,
            "volatilidad": 0.0
        })
        return resultado
    
    idx_mejor = rend_validos.idxmax()
    idx_peor = rend_validos.idxmin()
    
    resultado["rendimiento_total"] = rend_validos.sum()
    resultado["rendimiento_promedio"] = rend_validos.mean()
    resultado["mejor_dia"] = (idx_mejor, rend_validos.loc[idx_mejor])
    resultado["peor_dia"] = (idx_peor, rend_validos.loc[idx_peor])
    resultado["dias_positivos"] = int((rend_validos > 0).sum())
    resultado["dias_negativos"] = int((rend_validos < 0).sum())
    resultado["volatilidad"] = rend_validos.std()
    
    return resultado
````

````python
def media_movil(precios: pd.Series, ventana: int) -> pd.Series:
    """
    Calcula la media móvil simple (SMA).
    """
    return precios.rolling(window=ventana).mean()
````

````python
def bandas_bollinger(precios: pd.Series, ventana: int = 20, num_std: int = 2) -> Dict:
    """
    Calcula las Bandas de Bollinger.
    """
    sma = precios.rolling(window=ventana).mean()
    std = precios.rolling(window=ventana).std()
    
    resultado = {
        "banda_superior": sma + num_std * std,
        "banda_media": sma,
        "banda_inferior": sma - num_std * std
    }
    
    return resultado
````

````python
def detectar_maximos_minimos(precios: pd.Series, ventana: int = 5) -> Dict:
    """
    Detecta máximos y mínimos locales.
    Un máximo local: precio mayor o igual que los 'ventana' días anteriores y posteriores.
    Un mínimo local: precio menor o igual que los 'ventana' días anteriores y posteriores.
    """
    resultado = {
        "maximos": None,
        "minimos": None
    }
    
    # Ventana centrada de tamaño (2*ventana + 1): 'ventana' días antes, el día actual, 'ventana' días después
    max_movil = precios.rolling(window=2 * ventana + 1, center=True).max()
    min_movil = precios.rolling(window=2 * ventana + 1, center=True).min()
    
    es_maximo = precios == max_movil
    es_minimo = precios == min_movil
    
    resultado["maximos"] = precios[es_maximo]
    resultado["minimos"] = precios[es_minimo]
    
    return resultado
````

````python
def clasificar_tendencia(precios: pd.Series, ventana: int = 10) -> str:
    """
    Clasifica la tendencia actual.
    """
    ma = precios.rolling(window=ventana).mean()
    
    # Si no hay suficientes datos para comparar dos valores de la MA, no hay tendencia clara
    if ma.dropna().shape[0] < 2:
        return "LATERAL"
    
    precio_actual = precios.iloc[-1]
    ma_actual = ma.iloc[-1]
    ma_anterior = ma.iloc[-2]
    
    ma_subiendo = ma_actual > ma_anterior
    ma_bajando = ma_actual < ma_anterior
    
    if precio_actual > ma_actual and ma_subiendo:
        return "ALCISTA"
    elif precio_actual < ma_actual and ma_bajando:
        return "BAJISTA"
    else:
        return "LATERAL"
````

````python
def generar_senales_trading(precios: pd.Series, ma_corta: int = 5, ma_larga: int = 20) -> pd.Series:
    """
    Genera señales de compra/venta basadas en cruces de medias móviles.
    """
    mc = precios.rolling(window=ma_corta).mean()
    ml = precios.rolling(window=ma_larga).mean()
    
    # Diferencia entre medias: positiva si la corta está arriba de la larga
    diferencia = mc - ml
    diferencia_anterior = diferencia.shift(1)
    
    # Cruce hacia arriba: antes la corta estaba abajo (o igual), ahora está arriba
    cruce_arriba = (diferencia > 0) & (diferencia_anterior <= 0)
    # Cruce hacia abajo: antes la corta estaba arriba (o igual), ahora está abajo
    cruce_abajo = (diferencia < 0) & (diferencia_anterior >= 0)
    
    senales = pd.Series("MANTENER", index=precios.index)
    senales[cruce_arriba] = "COMPRA"
    senales[cruce_abajo] = "VENTA"
    
    # Donde alguna de las medias móviles no está definida aún, no puede haber señal válida
    senales[diferencia.isna() | diferencia_anterior.isna()] = "MANTENER"
    
    return senales
````

````python
def alertas_precio(precios: pd.Series, umbral_cambio: float = 5.0) -> List[Dict]:
    """
    Genera alertas cuando hay cambios significativos.
    """
    alertas = []
    
    rendimientos = calcular_rendimientos(precios)
    mascara = rendimientos.abs() > umbral_cambio
    
    for fecha, cambio in rendimientos[mascara].items():
        tipo = "SUBIDA" if cambio > 0 else "CAIDA"
        fecha_str = fecha.strftime('%Y-%m-%d') if hasattr(fecha, 'strftime') else str(fecha)
        alertas.append({"fecha": fecha_str, "tipo": tipo, "cambio": cambio})
    
    return alertas
````

````python
def clasificar_volatilidad(rendimientos: pd.Series) -> str:
    """
    Clasifica el nivel de volatilidad del activo.
    """
    std = rendimientos.dropna().std()
    
    if pd.isna(std):
        return "BAJA"
    elif std < 1:
        return "BAJA"
    elif std < 3:
        return "MEDIA"
    elif std < 5:
        return "ALTA"
    else:
        return "MUY ALTA"
````

````python
def generar_reporte_completo(precios: pd.Series, nombre_accion: str) -> Dict:
    """
    Genera un reporte completo de análisis.
    """
    rendimientos = calcular_rendimientos(precios)
    stats = estadisticas_basicas(precios)
    analisis_rend = analisis_rendimientos(rendimientos)
    tendencia = clasificar_tendencia(precios)
    volatilidad = clasificar_volatilidad(rendimientos)
    senales = generar_senales_trading(precios)
    senal_actual = senales.iloc[-1]
    alertas = alertas_precio(precios)
    
    fecha_inicio = precios.index[0]
    fecha_fin = precios.index[-1]
    
    reporte = {
        "nombre": nombre_accion,
        "periodo": {
            "inicio": str(fecha_inicio.date()) if hasattr(fecha_inicio, 'date') else str(fecha_inicio),
            "fin": str(fecha_fin.date()) if hasattr(fecha_fin, 'date') else str(fecha_fin),
            "dias": len(precios)
        },
        "estadisticas": stats,
        "rendimientos": analisis_rend,
        "tendencia": tendencia,
        "volatilidad": volatilidad,
        "senal_actual": senal_actual,
        "alertas_recientes": alertas
    }
    
    return reporte
````

````python
# Simular 60 días de precios de una acción
np.random.seed(42)  # Para reproducibilidad

# Generar fechas
fechas = pd.date_range(start='2024-01-01', periods=60, freq='B')  # B = días hábiles

# Generar precios con tendencia alcista y volatilidad
precio_inicial = 100
rendimientos_simulados = np.random.normal(0.002, 0.02, 60)  # Media 0.2%, std 2%
precios_simulados = precio_inicial * np.cumprod(1 + rendimientos_simulados)

# Crear Serie de precios
PRECIOS_ACCION = pd.Series(
    precios_simulados.round(2),
    index=fechas,
    name='ACME Corp'
)

print("Precios de ACME Corp (primeros 10 días):")
print(PRECIOS_ACCION.head(10))
print(f"\nTotal de días: {len(PRECIOS_ACCION)}")
````

````python
# Datos adicionales para pruebas más completas
np.random.seed(123)

# Acción con alta volatilidad
rend_volatil = np.random.normal(0, 0.05, 60)  # 5% de volatilidad diaria
precios_volatil = 50 * np.cumprod(1 + rend_volatil)
ACCION_VOLATIL = pd.Series(
    precios_volatil.round(2),
    index=fechas,
    name='VolatilTech'
)

# Acción con tendencia bajista
rend_bajista = np.random.normal(-0.005, 0.015, 60)  # Tendencia negativa
precios_bajista = 200 * np.cumprod(1 + rend_bajista)
ACCION_BAJISTA = pd.Series(
    precios_bajista.round(2),
    index=fechas,
    name='DeclineCorp'
)

print("Acciones disponibles para análisis:")
print(f"1. ACME Corp - Precio actual: ${PRECIOS_ACCION.iloc[-1]:.2f}")
print(f"2. VolatilTech - Precio actual: ${ACCION_VOLATIL.iloc[-1]:.2f}")
print(f"3. DeclineCorp - Precio actual: ${ACCION_BAJISTA.iloc[-1]:.2f}")
````

## Funciones de Visualización

````python
def mostrar_reporte(reporte: Dict) -> None:
    """Muestra el reporte de forma legible."""
    print("=" * 70)
    print(f"           REPORTE DE ANÁLISIS: {reporte['nombre']}")
    print("=" * 70)
    
    # Período
    periodo = reporte.get('periodo', {})
    print(f"\nPERÍODO DE ANÁLISIS")
    print("-" * 40)
    print(f"Inicio: {periodo.get('inicio', 'N/A')}")
    print(f"Fin: {periodo.get('fin', 'N/A')}")
    print(f"Días analizados: {periodo.get('dias', 'N/A')}")
    
    # Estadísticas
    stats = reporte.get('estadisticas', {})
    print(f"\nESTADÍSTICAS DE PRECIO")
    print("-" * 40)
    print(f"Precio actual:  ${stats.get('precio_actual', 0):,.2f}")
    print(f"Precio mínimo:  ${stats.get('precio_minimo', 0):,.2f}")
    print(f"Precio máximo:  ${stats.get('precio_maximo', 0):,.2f}")
    print(f"Precio promedio: ${stats.get('precio_promedio', 0):,.2f}")
    
    # Rendimientos
    rend = reporte.get('rendimientos', {})
    print(f"\nRENDIMIENTO")
    print("-" * 40)
    print(f"Rendimiento total: {rend.get('rendimiento_total', 0):+.2f}%")
    print(f"Rendimiento promedio diario: {rend.get('rendimiento_promedio', 0):+.3f}%")
    if rend.get('mejor_dia'):
        fecha_mejor = rend['mejor_dia'][0]
        fecha_mejor_str = fecha_mejor.strftime('%Y-%m-%d') if hasattr(fecha_mejor, 'strftime') else str(fecha_mejor)
        print(f"Mejor día: {fecha_mejor_str} ({rend['mejor_dia'][1]:+.2f}%)")
    if rend.get('peor_dia'):
        fecha_peor = rend['peor_dia'][0]
        fecha_peor_str = fecha_peor.strftime('%Y-%m-%d') if hasattr(fecha_peor, 'strftime') else str(fecha_peor)
        print(f"Peor día: {fecha_peor_str} ({rend['peor_dia'][1]:+.2f}%)")
    print(f"Días positivos: {rend.get('dias_positivos', 0)}")
    print(f"Días negativos: {rend.get('dias_negativos', 0)}")
    
    # Indicadores
    print(f"\nINDICADORES")
    print("-" * 40)
    print(f"Tendencia: {reporte.get('tendencia', 'N/A')}")
    print(f"Volatilidad: {reporte.get('volatilidad', 'N/A')}")
    print(f"Señal actual: {reporte.get('senal_actual', 'N/A')}")
    
    # Alertas
    alertas = reporte.get('alertas_recientes', [])
    if alertas:
        print(f"\nALERTAS RECIENTES")
        print("-" * 40)
        for alerta in alertas[-5:]:  # Últimas 5
            simbolo = "[+]" if alerta['tipo'] == 'SUBIDA' else "[-]"
            print(f"{simbolo} {alerta['fecha']}: {alerta['tipo']} de {alerta['cambio']:+.2f}%")
    
    print("\n" + "=" * 70)
````

````python
def visualizar_precios_texto(precios: pd.Series, ancho: int = 50) -> None:
    """Visualización simple de precios en texto (ASCII chart)."""
    min_precio = precios.min()
    max_precio = precios.max()
    rango = max_precio - min_precio
    
    print(f"\nGráfico de precios: {precios.name}")
    print(f"Max: ${max_precio:.2f}")
    print("-" * (ancho + 10))
    
    # Mostrar cada 3 días para no saturar
    for fecha, precio in precios.iloc[::3].items():
        posicion = int((precio - min_precio) / rango * ancho) if rango > 0 else ancho // 2
        barra = " " * posicion + "#"
        fecha_str = fecha.strftime('%m/%d') if hasattr(fecha, 'strftime') else str(fecha)[:5]
        print(f"{fecha_str} |{barra}")
    
    print("-" * (ancho + 10))
    print(f"Min: ${min_precio:.2f}")
````

### 1. RSI (Relative Strength Index)

````python
def calcular_rsi(precios: pd.Series, periodos: int = 14) -> pd.Series:
    """
    Calcula el RSI (Relative Strength Index).
    
    RSI = 100 - (100 / (1 + RS))
    RS = Promedio de ganancias / Promedio de pérdidas
    
    Interpretación:
    - RSI > 70: Sobrecomprado (posible caída)
    - RSI < 30: Sobrevendido (posible subida)
    """
    delta = precios.diff()
    
    # Separar ganancias (cambios positivos) y pérdidas (cambios negativos, en valor absoluto)
    ganancias = delta.clip(lower=0)
    perdidas = -delta.clip(upper=0)
    
    avg_ganancia = ganancias.rolling(window=periodos).mean()
    avg_perdida = perdidas.rolling(window=periodos).mean()
    
    rs = avg_ganancia / avg_perdida
    rsi = 100 - (100 / (1 + rs))
    
    # Casos especiales: si no hubo pérdidas, el RSI es 100 (todo ganancia)
    rsi[avg_perdida == 0] = 100
    # Si no hubo ni ganancias ni pérdidas (precio constante), RSI neutral = 50
    rsi[(avg_ganancia == 0) & (avg_perdida == 0)] = 50
    
    return rsi
````

### 2. Backtesting simple

````python
def backtest_estrategia(precios: pd.Series, senales: pd.Series, capital_inicial: float = 10000) -> Dict:
    """
    Simula la estrategia de trading y calcula rendimiento.
    """
    capital = capital_inicial
    acciones_en_cartera = 0.0
    num_operaciones = 0
    operaciones_ganadoras = 0
    precio_compra = None
    
    for fecha in precios.index:
        precio_actual = precios.loc[fecha]
        senal = senales.loc[fecha]
        
        if senal == "COMPRA" and acciones_en_cartera == 0:
            # Invertimos todo el capital disponible
            acciones_en_cartera = capital / precio_actual
            capital = 0.0
            precio_compra = precio_actual
            num_operaciones += 1
        elif senal == "VENTA" and acciones_en_cartera > 0:
            # Vendemos toda la posición
            capital = acciones_en_cartera * precio_actual
            if precio_actual > precio_compra:
                operaciones_ganadoras += 1
            acciones_en_cartera = 0.0
            precio_compra = None
    
    # Si terminamos con acciones en cartera (sin vender), las valoramos al último precio disponible
    if acciones_en_cartera > 0:
        capital = acciones_en_cartera * precios.iloc[-1]
    
    rendimiento_total = ((capital - capital_inicial) / capital_inicial) * 100
    
    return {
        "capital_final": capital,
        "rendimiento_total": rendimiento_total,
        "num_operaciones": num_operaciones,
        "operaciones_ganadoras": operaciones_ganadoras
    }
````

Prueba del backtesting de la estrategia

````python
# Prueba del backtesting
print("-- Backtesting de la estrategia de cruce de medias --")
senales_backtest = generar_senales_trading(PRECIOS_ACCION)
resultado_backtest = backtest_estrategia(PRECIOS_ACCION, senales_backtest, capital_inicial=10000)

print(f"Capital inicial:         $10,000.00")
print(f"Capital final:           ${resultado_backtest['capital_final']:,.2f}")
print(f"Rendimiento total:       {resultado_backtest['rendimiento_total']:+.2f}%")
print(f"Número de operaciones:   {resultado_backtest['num_operaciones']}")
print(f"Operaciones ganadoras:   {resultado_backtest['operaciones_ganadoras']}")
````
