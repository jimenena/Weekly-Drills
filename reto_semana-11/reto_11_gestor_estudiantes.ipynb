````python
def info_general(df_estudiantes: pd.DataFrame, df_calificaciones: pd.DataFrame) -> Dict:
    """
    Genera información general del sistema.
    """
    resultado = {
        "total_estudiantes": len(df_estudiantes),
        "total_registros_calif": len(df_calificaciones),
        "semestres": sorted(df_estudiantes['semestre'].unique().tolist()),
        "materias_con_registros": df_calificaciones['materia_id'].nunique()
    }
    
    return resultado
````

````python
def validar_datos(df_calificaciones: pd.DataFrame) -> Dict:
    """
    Valida la integridad de los datos.
    """
    columnas_calif = ['parcial_1', 'parcial_2', 'final']
    
    # Filas con al menos un valor nulo en las columnas de calificación
    registros_con_nulos = df_calificaciones[columnas_calif].isna().any(axis=1).sum()
    
    # Calificaciones fuera del rango válido [0, 10]
    fuera_rango_mask = pd.Series(False, index=df_calificaciones.index)
    for col in columnas_calif:
        fuera_rango_mask |= (df_calificaciones[col] < 0) | (df_calificaciones[col] > 10)
    calificaciones_fuera_rango = int(fuera_rango_mask.sum())
    
    datos_validos = (registros_con_nulos == 0) and (calificaciones_fuera_rango == 0)
    
    resultado = {
        "registros_con_nulos": int(registros_con_nulos),
        "calificaciones_fuera_rango": calificaciones_fuera_rango,
        "datos_validos": bool(datos_validos)
    }
    
    return resultado
````

````python
def buscar_estudiante(df_estudiantes: pd.DataFrame, criterio: str, valor: str) -> pd.DataFrame:
    """
    Busca estudiantes por diferentes criterios.
    """
    if criterio == 'boleta':
        return df_estudiantes[df_estudiantes['boleta'] == valor]
    elif criterio == 'nombre':
        # Búsqueda parcial, sin distinguir mayúsculas/minúsculas
        return df_estudiantes[df_estudiantes['nombre'].str.contains(valor, case=False, na=False)]
    elif criterio == 'semestre':
        return df_estudiantes[df_estudiantes['semestre'] == int(valor)]
    else:
        return df_estudiantes.iloc[0:0]  # DataFrame vacío con las mismas columnas
````

````python
def _promedio_por_materia(df_calificaciones: pd.DataFrame) -> pd.Series:
    """Promedio de los 3 parciales (parcial_1, parcial_2, final) para cada registro."""
    return df_calificaciones[['parcial_1', 'parcial_2', 'final']].mean(axis=1)


def obtener_kardex(boleta: str, df_estudiantes: pd.DataFrame,
                   df_calificaciones: pd.DataFrame, df_materias: pd.DataFrame) -> Dict:
    """
    Obtiene el kardex completo de un estudiante.
    """
    resultado = {
        "estudiante": None,
        "materias": None,
        "promedio_general": None,
        "creditos_cursados": None,
        "materias_aprobadas": None,
        "materias_reprobadas": None
    }
    
    # 1. Obtener datos del estudiante
    est_df = df_estudiantes[df_estudiantes['boleta'] == boleta]
    if est_df.empty:
        return resultado
    
    resultado["estudiante"] = est_df.iloc[0].to_dict()
    
    # 2. Filtrar calificaciones del estudiante
    calif_est = df_calificaciones[df_calificaciones['boleta'] == boleta].copy()
    
    if calif_est.empty:
        resultado["promedio_general"] = 0.0
        resultado["creditos_cursados"] = 0
        resultado["materias_aprobadas"] = 0
        resultado["materias_reprobadas"] = 0
        resultado["materias"] = calif_est
        return resultado
    
    # Promedio de los 3 parciales por materia cursada
    calif_est['promedio'] = _promedio_por_materia(calif_est)
    
    # 3. Unir con nombres de materias
    materias_completas = calif_est.merge(
        df_materias[['materia_id', 'nombre', 'creditos']],
        on='materia_id',
        how='left'
    )
    
    columnas_finales = ['materia_id', 'nombre', 'parcial_1', 'parcial_2', 'final', 'promedio', 'creditos']
    resultado["materias"] = materias_completas[columnas_finales]
    
    # 4. Calcular promedio general (promedio de los promedios de cada materia)
    resultado["promedio_general"] = calif_est['promedio'].mean()
    resultado["creditos_cursados"] = int(materias_completas['creditos'].sum())
    
    # 5. Contar aprobadas (promedio >= 6) y reprobadas
    resultado["materias_aprobadas"] = int((calif_est['promedio'] >= 6).sum())
    resultado["materias_reprobadas"] = int((calif_est['promedio'] < 6).sum())
    
    return resultado
````

````python
def filtrar_por_rendimiento(df_calificaciones: pd.DataFrame,
                            df_estudiantes: pd.DataFrame,
                            min_promedio: float = None,
                            max_promedio: float = None) -> pd.DataFrame:
    """
    Filtra estudiantes por rango de promedio.
    """
    # 1. Calcular promedio por estudiante (promedio de todas sus materias)
    df_calif = df_calificaciones.copy()
    df_calif['promedio_registro'] = _promedio_por_materia(df_calif)
    
    promedios = df_calif.groupby('boleta')['promedio_registro'].mean().reset_index()
    promedios.columns = ['boleta', 'promedio']
    
    # 2. Aplicar filtros según min_promedio y max_promedio
    if min_promedio is not None:
        promedios = promedios[promedios['promedio'] >= min_promedio]
    if max_promedio is not None:
        promedios = promedios[promedios['promedio'] <= max_promedio]
    
    # 3. Unir con datos de estudiantes
    resultado = promedios.merge(df_estudiantes, on='boleta', how='left')
    
    return resultado
````

````python
def calcular_promedio_materia(df_calificaciones: pd.DataFrame, materia_id: str) -> Dict:
    """
    Calcula estadísticas de una materia.
    """
    resultado = {
        "materia": materia_id,
        "inscritos": None,
        "promedio_parcial1": None,
        "promedio_parcial2": None,
        "promedio_final": None,
        "promedio_general": None,
        "tasa_aprobacion": None,
        "calificacion_maxima": None,
        "calificacion_minima": None
    }
    
    datos = df_calificaciones[df_calificaciones['materia_id'] == materia_id].copy()
    
    if datos.empty:
        resultado["inscritos"] = 0
        return resultado
    
    datos['promedio'] = _promedio_por_materia(datos)
    
    resultado["inscritos"] = len(datos)
    resultado["promedio_parcial1"] = datos['parcial_1'].mean()
    resultado["promedio_parcial2"] = datos['parcial_2'].mean()
    resultado["promedio_final"] = datos['final'].mean()
    resultado["promedio_general"] = datos['promedio'].mean()
    resultado["tasa_aprobacion"] = (datos['promedio'] >= 6).sum() / len(datos) * 100
    resultado["calificacion_maxima"] = datos['promedio'].max()
    resultado["calificacion_minima"] = datos['promedio'].min()
    
    return resultado
````

````python
def ranking_estudiantes(df_calificaciones: pd.DataFrame,
                        df_estudiantes: pd.DataFrame,
                        top_n: int = 10) -> pd.DataFrame:
    """
    Genera ranking de mejores estudiantes por promedio.
    """
    # 1. Calcular promedio por estudiante (todas sus materias)
    df_calif = df_calificaciones.copy()
    df_calif['promedio_registro'] = _promedio_por_materia(df_calif)
    
    promedios = df_calif.groupby('boleta')['promedio_registro'].mean().reset_index()
    promedios.columns = ['boleta', 'promedio']
    
    # 4. Agregar datos del estudiante (nombre, semestre)
    ranking = promedios.merge(df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta', how='left')
    
    # 2. Ordenar descendente y 3. Tomar top_n
    ranking = ranking.sort_values('promedio', ascending=False).head(top_n).reset_index(drop=True)
    ranking.insert(0, 'posicion', range(1, len(ranking) + 1))
    ranking = ranking[['posicion', 'nombre', 'semestre', 'promedio']]
    ranking['promedio'] = ranking['promedio'].round(2)
    
    return ranking
````

````python
def estadisticas_por_semestre(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame) -> pd.DataFrame:
    """
    Calcula estadísticas agrupadas por semestre.
    """
    # 1. Calcular promedio por estudiante
    df_calif = df_calificaciones.copy()
    df_calif['promedio_registro'] = _promedio_por_materia(df_calif)
    
    promedios = df_calif.groupby('boleta')['promedio_registro'].mean().reset_index()
    promedios.columns = ['boleta', 'promedio']
    
    # 2. Unir con semestre del estudiante
    datos = promedios.merge(df_estudiantes[['boleta', 'semestre']], on='boleta', how='left')
    
    # 3. Agrupar por semestre y calcular estadísticas
    def tasa_aprobacion(serie):
        return (serie >= 6).sum() / len(serie) * 100
    
    agrupado = datos.groupby('semestre').agg(
        Estudiantes=('boleta', 'count'),
        Promedio=('promedio', 'mean'),
        Mejor=('promedio', 'max'),
        Peor=('promedio', 'min')
    )
    agrupado['Tasa_Aprob'] = datos.groupby('semestre')['promedio'].apply(tasa_aprobacion)
    
    agrupado['Promedio'] = agrupado['Promedio'].round(2)
    agrupado['Mejor'] = agrupado['Mejor'].round(2)
    agrupado['Peor'] = agrupado['Peor'].round(2)
    agrupado['Tasa_Aprob'] = agrupado['Tasa_Aprob'].round(1)
    agrupado.index.name = 'Semestre'
    
    return agrupado
````

````python
def identificar_estudiantes_riesgo(df_calificaciones: pd.DataFrame,
                                   df_estudiantes: pd.DataFrame,
                                   umbral_promedio: float = 7.0,
                                   max_reprobadas: int = 2) -> pd.DataFrame:
    """
    Identifica estudiantes en riesgo académico.
    """
    # 1. Calcular promedio general por estudiante
    df_calif = df_calificaciones.copy()
    df_calif['promedio_registro'] = _promedio_por_materia(df_calif)
    df_calif['reprobada'] = df_calif['promedio_registro'] < 6
    
    # 2. Contar materias reprobadas por estudiante
    resumen = df_calif.groupby('boleta').agg(
        promedio=('promedio_registro', 'mean'),
        reprobadas=('reprobada', 'sum')
    ).reset_index()
    
    # 3. Identificar quienes cumplen criterios de riesgo
    bajo_promedio = resumen['promedio'] < umbral_promedio
    muchas_reprobadas = resumen['reprobadas'] > max_reprobadas
    
    en_riesgo = resumen[bajo_promedio | muchas_reprobadas].copy()
    
    # 4. Agregar columna 'motivo'
    def determinar_motivo(row):
        es_bajo = row['promedio'] < umbral_promedio
        son_muchas = row['reprobadas'] > max_reprobadas
        if es_bajo and son_muchas:
            return "Ambos"
        elif son_muchas:
            return "Materias reprobadas"
        else:
            return "Bajo promedio"
    
    en_riesgo['motivo'] = en_riesgo.apply(determinar_motivo, axis=1)
    
    resultado = en_riesgo.merge(df_estudiantes[['boleta', 'nombre']], on='boleta', how='left')
    resultado['promedio'] = resultado['promedio'].round(2)
    resultado['reprobadas'] = resultado['reprobadas'].astype(int)
    
    resultado = resultado[['boleta', 'nombre', 'promedio', 'reprobadas', 'motivo']]
    resultado = resultado.sort_values('promedio').reset_index(drop=True)
    
    return resultado
````

````python
def generar_reporte_academico(df_estudiantes: pd.DataFrame,
                              df_calificaciones: pd.DataFrame,
                              df_materias: pd.DataFrame) -> Dict:
    """
    Genera reporte académico completo.
    """
    reporte = {
        "resumen_general": {},
        "por_semestre": None,
        "por_materia": None,
        "mejores_estudiantes": None,
        "estudiantes_riesgo": None,
        "fecha_generacion": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    
    df_calif = df_calificaciones.copy()
    df_calif['promedio_registro'] = _promedio_por_materia(df_calif)
    
    promedio_global = df_calif['promedio_registro'].mean()
    tasa_aprobacion_global = (df_calif['promedio_registro'] >= 6).sum() / len(df_calif) * 100
    
    reporte["resumen_general"] = {
        "total_estudiantes": len(df_estudiantes),
        "promedio_global": promedio_global,
        "tasa_aprobacion": tasa_aprobacion_global
    }
    
    reporte["por_semestre"] = estadisticas_por_semestre(df_estudiantes, df_calificaciones)
    
    por_materia = []
    for materia_id in df_materias['materia_id']:
        stats = calcular_promedio_materia(df_calificaciones, materia_id)
        por_materia.append(stats)
    reporte["por_materia"] = pd.DataFrame(por_materia)
    
    reporte["mejores_estudiantes"] = ranking_estudiantes(df_calificaciones, df_estudiantes, top_n=5)
    reporte["estudiantes_riesgo"] = identificar_estudiantes_riesgo(df_calificaciones, df_estudiantes)
    
    return reporte
````

````python
def exportar_kardex(boleta: str, kardex: Dict, formato: str = 'csv') -> str:
    """
    Exporta el kardex de un estudiante a archivo.
    """
    # 1. Crear nombre de archivo con boleta y fecha
    fecha_str = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"kardex_{boleta}_{fecha_str}.{formato}"
    
    materias_df = kardex["materias"] if kardex.get("materias") is not None else pd.DataFrame()
    
    # 2. Exportar según formato
    if formato == 'csv':
        materias_df.to_csv(nombre_archivo, index=False)
    elif formato == 'json':
        export_data = {
            "estudiante": kardex.get("estudiante"),
            "materias": materias_df.to_dict(orient='records'),
            "promedio_general": kardex.get("promedio_general"),
            "creditos_cursados": kardex.get("creditos_cursados"),
            "materias_aprobadas": kardex.get("materias_aprobadas"),
            "materias_reprobadas": kardex.get("materias_reprobadas")
        }
        with open(nombre_archivo, 'w', encoding='utf-8') as f:
            json.dump(export_data, f, indent=2, ensure_ascii=False, default=str)
    
    return nombre_archivo
````

### 1. Predictor de riesgo

````python
def predecir_riesgo_proximo_semestre(df_calificaciones: pd.DataFrame,
                                     df_estudiantes: pd.DataFrame) -> pd.DataFrame:
    """
    Predice estudiantes que podrían estar en riesgo el próximo semestre
    basándose en tendencias de calificaciones.
    
    Criterios:
    - Tendencia decreciente en parciales (parcial_2 < parcial_1)
    - Calificación final menor que parcial 1
    
    Un estudiante se marca como candidato a riesgo si al menos el 50% de sus
    materias cursadas muestran alguno de estos patrones de tendencia negativa.
    """
    df_calif = df_calificaciones.copy()
    df_calif['tendencia_decreciente'] = df_calif['parcial_2'] < df_calif['parcial_1']
    df_calif['final_menor_p1'] = df_calif['final'] < df_calif['parcial_1']
    df_calif['en_riesgo_materia'] = df_calif['tendencia_decreciente'] | df_calif['final_menor_p1']
    
    resumen = df_calif.groupby('boleta').agg(
        materias_con_tendencia_negativa=('en_riesgo_materia', 'sum'),
        total_materias=('materia_id', 'count')
    ).reset_index()
    
    resumen['proporcion_riesgo'] = resumen['materias_con_tendencia_negativa'] / resumen['total_materias']
    
    # Umbral: al menos la mitad de las materias muestran tendencia negativa
    candidatos = resumen[resumen['proporcion_riesgo'] >= 0.5].copy()
    
    resultado = candidatos.merge(df_estudiantes[['boleta', 'nombre', 'semestre']], on='boleta', how='left')
    resultado = resultado.sort_values('proporcion_riesgo', ascending=False).reset_index(drop=True)
    resultado['proporcion_riesgo'] = (resultado['proporcion_riesgo'] * 100).round(1)
    resultado = resultado.rename(columns={'proporcion_riesgo': 'pct_materias_riesgo'})
    
    columnas = ['boleta', 'nombre', 'semestre', 'materias_con_tendencia_negativa', 'total_materias', 'pct_materias_riesgo']
    return resultado[columnas]
````

### 2. Comparador de rendimiento

````python
def comparar_estudiantes(boleta1: str, boleta2: str,
                         df_calificaciones: pd.DataFrame,
                         df_estudiantes: pd.DataFrame,
                         df_materias: pd.DataFrame) -> Dict:
    """
    Compara el rendimiento de dos estudiantes.
    """
    kardex1 = obtener_kardex(boleta1, df_estudiantes, df_calificaciones, df_materias)
    kardex2 = obtener_kardex(boleta2, df_estudiantes, df_calificaciones, df_materias)
    
    if kardex1['estudiante'] is None or kardex2['estudiante'] is None:
        return None
    
    materias1 = set(kardex1['materias']['materia_id']) if kardex1['materias'] is not None else set()
    materias2 = set(kardex2['materias']['materia_id']) if kardex2['materias'] is not None else set()
    materias_comunes = materias1 & materias2
    
    comparacion_materias = []
    if materias_comunes:
        m1 = kardex1['materias'].set_index('materia_id')
        m2 = kardex2['materias'].set_index('materia_id')
        for materia_id in materias_comunes:
            comparacion_materias.append({
                'materia_id': materia_id,
                'nombre': m1.loc[materia_id, 'nombre'],
                f'promedio_{boleta1}': m1.loc[materia_id, 'promedio'],
                f'promedio_{boleta2}': m2.loc[materia_id, 'promedio']
            })
    
    resultado = {
        'estudiante_1': {
            'boleta': boleta1,
            'nombre': kardex1['estudiante']['nombre'],
            'promedio_general': kardex1['promedio_general'],
            'materias_aprobadas': kardex1['materias_aprobadas'],
            'materias_reprobadas': kardex1['materias_reprobadas']
        },
        'estudiante_2': {
            'boleta': boleta2,
            'nombre': kardex2['estudiante']['nombre'],
            'promedio_general': kardex2['promedio_general'],
            'materias_aprobadas': kardex2['materias_aprobadas'],
            'materias_reprobadas': kardex2['materias_reprobadas']
        },
        'materias_comunes': pd.DataFrame(comparacion_materias),
        'diferencia_promedio': kardex1['promedio_general'] - kardex2['promedio_general'],
        'mejor_promedio': boleta1 if kardex1['promedio_general'] > kardex2['promedio_general'] else boleta2
    }
    
    return resultado
````
